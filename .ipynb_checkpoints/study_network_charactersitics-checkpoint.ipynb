```python
data_relevant = pd.read_csv("data_relevant_mars_30", sep=",")
data_relevant
```

```python
def introduce_trade_balance(df,year):
    '''adding column capturing trade balance between two countries
    
    Parameters:
    ---
    df: 
        dataframe with import export data, here: data_relevant
    year: int
        year for which the trade balance is calculated
        
    Return:
    ---
    df: 
        dataframe with extra column
    '''
    df_temp = df.copy()
    trade_balances = []   
    for export_row in df_temp.iterrows():
        exp_value = export_row[1]["v{}".format(year)]
        imp_row = df_temp[(df_temp["importer"] == export_row[1]["exporter"]) & (df_temp["exporter"] == export_row[1]["importer"])]
        if len(imp_row) != 0:
            imp_value = imp_row.iloc[0]["v{}".format(year)]
        else:
            imp_value = np.nan
        trade_balance = exp_value - imp_value
        trade_balances.append(trade_balance)
    df_temp["trade_balance_{}".format(year)] = trade_balances

    return df_temp
```

```python
for year in [2018, 2019, 2020]:
    data_relevant = introduce_trade_balance(data_relevant, year)
    print(len(data_relevant[~data_relevant["trade_balance_{}".format(year)].isna()]))
```

```python
Graph_2018 = nx.DiGraph()
Graph_2018 = nx.from_pandas_edgelist(data_relevant[["exporter", "importer", "v2018", "relative_volume_2018", "trade_balance_2018"]], 
                                     source = 'exporter',
                                     target = 'importer',
                                     edge_attr = True,
                                     create_using = nx.DiGraph())

Graph_2020 = nx.DiGraph()
Graph_2020 = nx.from_pandas_edgelist(data_relevant[["exporter", "importer", "v2020", "relative_volume_2020", "trade_balance_2018"]], 
                                     source = 'exporter',
                                     target = 'importer',
                                     edge_attr = True,
                                     create_using = nx.DiGraph())
```

```python
for graph in [Graph_2018, Graph_2020]:
    print("\n")
    print("Reciprocity: ", nx.reciprocity(graph))
    print("Strongly connected ? ", nx.is_strongly_connected((graph)))
    print( "__________now to undirected_________",)
    print("Transitivity: ", nx.transitivity(nx.DiGraph.to_undirected(graph))) #not for multidigraph
    print("Average clustering : ", nx.average_clustering(nx.DiGraph.to_undirected(graph)))
    print("Density : ", nx.density(nx.DiGraph.to_undirected(graph)))
    print("Top 10 Eigenvector centrality : ", pd.DataFrame.from_dict(nx.eigenvector_centrality(nx.DiGraph.to_undirected(graph)),orient='index').sort_values(by = [0], ascending= False).head(10))


```
